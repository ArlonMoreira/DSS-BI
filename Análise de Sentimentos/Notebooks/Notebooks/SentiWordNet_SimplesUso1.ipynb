SentiWordNet pode ser importada usando este comando:

```python
 from nltk.corpus import sentiwordnet as swn
```

Para usar a sentwordNet é preciso fazer o download da WordNet

```python
import nltk
nltk.download('wordnet')
```

Saber em quais synsets a palavra que está. Usa-se a palvra e a função sintática. Por exemplo 'happy' é um adjetivo, portanto usa-se o parâmetro 'a'

```python
happy = swn.senti_synsets('happy', 'a')
print(list(happy))
```

```python
breakdown = swn.senti_synset('breakdown.n.03')
print(breakdown)
```

```python
happy = swn.senti_synset('happy.a.01')
print(happy)
```

```python

breakdown.pos_score()
```

```python
breakdown.neg_score()
```

```python
breakdown.obj_score()
```

Quais são as palavras que fazem parte do Synset do termo'slow'?

```python
list(swn.senti_synsets('slow'))
```

Quais as palavras que fazem parte do Synset do termo 'happy'?

```python
happy = swn.senti_synsets('happy', 'a')
list(happy)
```

Como listar todos os synsets existentes no lexico?

```python
all = swn.all_senti_synsets()
list(all)
```
